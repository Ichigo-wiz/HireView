```python
def main():
    print("          -:   Welcome to Hire View   :-")
    scanner=Scanning()
    apps=Applicant()
    #Initialising
    while True:
        print("::::::::::::::::::::::::::::::::::::::::::::::::::::::::")
        print()
        print("Enter 1 to scan resume")
        print("Enter 2 to see all applicants")
        print("Enter 3 to view skills of a particular applicant")
        print("Enter 4 to view top 5 applicants")
        print("Enter 5 to search by a particular skill")
        print("Enter 6 to exit")
        choice=int(input("Enter your choice - "))
        print()
        match choice:
            case 1:
                scanner.scan()
                continue
            case 2:
                apps.see_names()
                continue
            case 3:
                apps.search_applicant()
                continue
            case 4:
                apps.top_five_applicants()
                continue
            case 5:
                apps.search_by_skill()
                continue
            case 6:
                print()
                print("          -:   Thank You for using Hire View   :-")
                print()
                break
            case _:
                print("Value not found")


#Scanning resumes
class Scanning:
    def __init__(self):
        self.app_name=None
        self.app_numb=None
        self.app_github=None
        self.app_linkedin=None
        self.app_mail=None
        self.app_skills=[]
        self.no_of_skills=None
    def scan(self):
        while True:
            print("::::::::::::::::::::::::::::::::::::::::::::::::::::::::")
            print()
            print("Enter 1 to Upload single Resume")
            print("Enter 2 to Upload a folder of resumes")
            print("Enter 3 to exit")
            choice=int(input("Enter your choice - "))
            print()
            match choice:
                case 1:
                    self.read_resume()
                    self.data_insertion()
                    return
                case 2:
                    self.scan_folder()
                    return
                case 3:
                    break
                case _:
                    print("Value not found")
    
    
    #Scanning single resume
    def read_resume(self):
        file_path=input("Enter File Path - ")
        if file_path.count(".pdf"):
            self.scan_pdf(file_path)
        elif file_path.count(".doc"):
            self.scan_word(file_path)
        else:
            print("File type not supported")
    #Scanning folders
    def scan_folder(self):
        folder_path = input("Enter Folder Path - ")
        try:
            for file in listdir(folder_path):
                file_path = path.join(folder_path, file)
                if path.isfile(file_path):
                    if file_path.count(".pdf"):
                        self.scan_pdf(file_path)
                        self.data_insertion()
                    elif file_path.count(".doc"):
                        self.scan_word(file_path)
                        self.data_insertion()
                    else:
                        raise Exception("File type not supported")
        except Exception as e:
            print(f"Error: {e}")
            
            
    #Scanning pdfs
    def scan_pdf(self,file_path):
        reader = PdfReader(file_path)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
        lines=text.splitlines()
        name=lines[0].strip()
        mail=numb=github=linkedin=""
        special_characters = "@!#$%&'*+/=?^_`{|}~.-"
        for line in lines:
            if "+91" in line:
                ind=line.index("+91")
                for char in range(ind+3,len(line)):
                    if line[char].isdigit():
                        numb+=line[char]
                    if len(numb)==10:
                        break
            if "mail.com" in line:
                ind=line.index("mail.com")
                left=right=ind
                
                while line[left]!=" " and left>0:
                    if line[left] not in special_characters and not line[left].isalnum():
                        break
                    left=left-1
                while line[right]!=" " and right<len(line)-1:
                    if line[right] not in special_characters and not line[right].isalnum():
                        break
                    right=right+1
                mail=line[left:right+1]
                if mail.startswith("⌢pe"):
                    mail=mail[3:]
            if "linkedin.com" in line:
                end=ind=line.index("linkedin.com")
                for index in range(ind,len(line)):
                    if line[index]==" ":
                        break 
                    end+=1
                linkedin=line[ind:end+1]
            if "github.com" in line:
                end=ind=line.index("github.com")
                for index in range(ind,len(line)):
                    if line[index]==" ":
                        break
                    end+=1
                github=line[ind:end+1]
        skills=[]
        text=text.lower()
        for word in sde_skills:
            ind=text.find(word)
            if(ind!=-1):
                skills.append(word)
        self.app_name=name
        self.app_numb=numb
        self.app_mail=mail
        self.app_github=github
        self.app_linkedin=linkedin
        self.app_skills="  ,  ".join(skills)
        self.no_of_skills=len(skills)


#Scanning word files
    def scan_word(self,f_path):
        doc = docx.Document(f_path)
        mail=numb=github=linkedin=""
        skills=[]
        content = doc.paragraphs
        lines = [para.text for para in content]
        name=lines[0]
        special_characters = "@!#$%&'*+/=?^_`{|}~.-"
        for line in lines:
            if "+91" in line:
                ind=line.index("+91")
                for char in range(ind+3,len(line)):
                    if line[char].isdigit():
                        numb+=line[char]
                    if len(numb)==10:
                        break
            if "mail.com" in line:
                ind=line.index("mail.com")
                left=right=ind
                
                while line[left]!=" " and left>0:
                    if line[left] not in special_characters and not line[left].isalnum():
                        break
                    left=left-1
                while line[right]!=" " and right<len(line)-1:
                    if line[right] not in special_characters and not line[right].isalnum():
                        break
                    right=right+1
                mail=line[left:right+1]
                if mail.startswith("⌢pe"):
                    mail=mail[3:]
            if "linkedin.com" in line:
                end=ind=line.index("linkedin.com")
                for index in range(ind,len(line)):
                    if line[index]==" ":
                        break
                    else:
                        pass    
                    end+=1
            
                linkedin=line[ind:end+1]
            if "github.com" in line:
                end=ind=line.index("github.com")
                for index in range(ind,len(line)):
                    if line[index]==" ":
                        break
                    end+=1
                github=line[ind:end+1]
        cont="\n".join(lines).lower()
        for word in sde_skills:
            if word in cont:
                skills.append(word)
        self.app_name=name
        self.app_numb=numb
        self.app_mail=mail
        self.app_github=github
        self.app_linkedin=linkedin
        self.app_skills="  ,  ".join(skills)
        self.no_of_skills=len(skills)
        
        
#Inserting Data
    def data_insertion(self):
        try:
            conn = sql.connect(host="localhost", user="root", passwd="",database="hireview")
        except sql.Error as e:
            print(f"Database connection failed: {e}")
        cur=conn.cursor()
        query="INSERT INTO applicant_details (Name,Mobile,Email,Github,LinkedIn) VALUES (%s,%s,%s,%s,%s)"
        values=(self.app_name,self.app_numb,self.app_mail,self.app_github,self.app_linkedin)
        cur.execute(query,values)
        conn.commit()
        select_query="SELECT Applicant_id FROM applicant_details WHERE Name= %s ORDER BY Applicant_id DESC"
        value1=(self.app_name,)
        cur.execute(select_query,value1)
        id=cur.fetchall()[0][0]
        query="INSERT INTO skills (Applicant_id,Skills,Number_of_Skills) VALUES (%s,%s,%s)"
        values=(id,self.app_skills,self.no_of_skills)
        cur.execute(query,values)
        conn.commit()
        return


#Applicant Info
class Applicant:
    applicants=[]
    def __init__(self):
        self.app_id=None
        self.app_name=None


    
    def see_names(self):
        try:
            conn = sql.connect(host="localhost", user="root", passwd="",database="hireview")
        except sql.Error as e:
            print(f"Database connection failed: {e}")
        cur=conn.cursor()
        query="SELECT Applicant_id,Name FROM applicant_details"
        cur.execute(query)
        data=DataFrame(cur.fetchall(),columns=["Applicant_id","Name"])
        print(data.to_string(index=False))

    
    def search_applicant(self):
        try:
            conn = sql.connect(host="localhost", user="root", passwd="",database="hireview")
        except sql.Error as e:
            print(f"Database connection failed: {e}")
        cur=conn.cursor()
        query="SELECT Applicant_id,Skills from skills"
        cur.execute(query)
        data=DataFrame(cur.fetchall(),columns=["Applicant_id","Skills"])
        id=int(input("Enter ID of Applicant - "))
        if id not in data["Applicant_id"].values:
            print("Invalid Applicant ID")
        else:
            print(data[data["Applicant_id"]==id]["Skills"].values[0])

    def top_five_applicants(self):
        try:
            conn = sql.connect(host="localhost", user="root", passwd="",database="hireview")
        except sql.Error as e:
            print(f"Database connection failed: {e}")
        query="SELECT applicant_details.Applicant_id,applicant_details.Name,skills.Number_Of_Skills,Skills from applicant_details INNER JOIN skills on applicant_details.Applicant_id=skills.Applicant_id ORDER BY skills.Number_Of_Skills DESC"
        cur=conn.cursor()
        cur.execute(query)
        data=DataFrame(cur.fetchall(),columns=["Applicant_id","Name","Number_of_Skills","Skills"])
        data=data[:5]
        print(data.to_string(index=False))
        extract_skills=array(data["Number_of_Skills"].values)
        names=array(data["Name"].values)
        subplot(2,1,1)
        pie(extract_skills,labels=names,colors=["red","blue","green","orange","yellow"],shadow=True,autopct="%1.1f%%")
        title("Top 5 Applicants")
        subplot(2,1,2)
        bar(names,extract_skills,color=["red","blue","green","orange","yellow"])
        title("Top 5 Applicants")
        tight_layout()
        show()
        
    def search_by_skill(self):
        try:
            conn = sql.connect(host="localhost", user="root", passwd="", database="hireview")
        except sql.Error as e:
            print(f"Database connection failed: {e}")
        cur = conn.cursor()
        query = "SELECT applicant_details.Applicant_id, applicant_details.Name, Skills FROM applicant_details INNER JOIN skills ON applicant_details.Applicant_id = skills.Applicant_id"
        cur.execute(query)
        data = DataFrame(cur.fetchall(), columns=["Applicant_id", "Name", "Skills"])
        cur.close()
        conn.close()
        skill = input("Enter skill you want to search for - ").strip().lower()
        data["Skills"] = data["Skills"].str.lower() 
        filtered_data = data[data["Skills"].str.contains(fr'\b{skill}\b', regex=True, na=False)]
        if not filtered_data.empty:
            print(filtered_data.iloc[:, :2].to_string(index=False))
        else:
            print(f"None of the applicants have '{skill}' as their skill.")
```
